# Principali Distribuzioni

## Uniforme


\begin{equation}
p(x) =\begin{cases}
		1, &\text{se } 0<x<1\\
        0, & \text{altrimenti}
	 \end{cases}
\end{equation}
 Una distribuzione molto semplice per cui tutte le variabili casuali tra $(0,1)$ sono ugualmente probabili.
 
## Bernoulli

Distribuzione di una variabile casuale che assume il valore $k=1$ (successo) con probabilità $p$ e il valore $k=0$ (insuccesso), altrimenti

\begin{equation}
 p(k) = p^k (1-p)^{1-k},
\end{equation}

che è solo una forma chiusa per dire $p(1)=p$, $p(0)= 1-p(1) = 1-p$.

## Binomiale

Probabilità di $k$ successi in $n$ tentativi con probabilità individuale di successo $p$,

\begin{equation}
p(k | n, p) =  \binom{n}{k}p^k (1-p)^{n-k},
\end{equation}
composta da a) il coefficiente binomiale $\binom{n}{k}$, che sono le possibili combinazioni di $k$ prelievi da $n$ elementi, b) $p^k$, la probabilità di $k$ successi, se ogni successo ha probabilità $p$, e c) $(1-p)^{n-k}$, la probabilità di $n-k$ insuccessi, se ogni insuccesso ha probabilità $1-p$.

## Geometrica
Considera un evento con probabilità di successo $p$. Qual è la probabilità di fallire $k$ volte prima che accada una volta? Seguendo la spiegazione della distribuzione binomiale, segue immediatamente che 
\begin{equation}
p(k|p) = (1-p)^k \,p
\end{equation}

Qual è la probabilità di successo entro $n$ tentativi o meno?

\begin{equation}
\begin{split}
p(k<n | p)& = p \sum_{k=0}^{n-1} (1-p)^k = p \frac{1-(1-p)^n}{1-(1-p)} = \\
& = 1 - (1-p)^n,
\end{split}
\end{equation}
 dove abbiamo usato l'espressione standard per una somma geometrica, $\sum_{k=0}^{n-1} r^k = (1-r^n) / (1-r)$.
 
## Variabili continue

Fino ad ora, abbiamo considerato solo probabilità di variabili casuali discrete. Vediamo come si potrebbe prendere il limite per variabili casuali continue e derivare la distribuzione esponenziale. 

Considera un piccolo intervallo di tempo $\Delta$, in ciascuno dei quali effettui un tentativo con probabilità di successo $p = \Delta \lambda$, con $\lambda$ costante. In un tempo $t$ avrai $n=t/\Delta$ tentativi. La probabilità di avere avuto un successo entro $n$ tentativi (cioè con $n$ o meno), è data dall'equazione sopra,
 \begin{equation}
 P\Big(k<n=\frac{t}{\Delta}\Big) = 1 - (1-p)^n = 1 - (1 - \lambda \Delta)^\frac{t}{\Delta}.
 \end{equation}
 Nel limite $\Delta \rightarrow 0$ (un tentativo in ogni istante di tempo) troviamo 
 \begin{equation}
 P(<t) = 1 - \text{exp}(-\lambda t),
 \end{equation}
 come una delle definizioni dell'esponenziale è $\text{exp}(x) = \lim_{n\rightarrow 0}(1+x n)^{1/n}$. Si noti che l'idea di un tentativo in ogni momento del tempo ha senso solo se ciascuno di questi tentativi ha una probabilità di successo infinitesimale, inversamente proporzionale alla durata infinitesimale del tentativo. Se i molti tentativi avessero avuto probabilità di successo finite, l'evento sarebbe accaduto con certezza in qualsiasi intervallo di tempo $\epsilon>0$. Questo ci lascia praticamente con un processo con un tasso costante $\lambda$ di occorrenza. Questo ci porta alla prossima distribuzione, e la prima continua: la
 
## Distribuzione esponenziale 
Probabilità che un evento con tasso $\lambda$ si verifichi al tempo $t$,
\begin{equation}
\text{Exp}(t|\lambda) = \frac{\text{d}}{\text{d}t} \left[ 1 - \text{exp}(-\lambda t) \right] = \lambda \text{exp}(-\lambda t).
\end{equation}
Questa è la distribuzione tipica per il tempo di decadimento radioattivo, anche se in quel contesto, spesso definiamo il tempo di dimezzamento $\tau_{1/2}$, dato da $P(<\tau_{1/2}|\lambda)=1/2$. Invertendo questa espressione si può facilmente trovare $\ln 2 = \lambda \tau_{1/2}$.


## Distribuzione di Poisson

Probabilità di contare $n$ eventi, se ci si aspettava che si verificassero $\mu$ eventi
  \begin{equation}
\text{Possion}(n|\mu) = \frac{1}{n!} \mu^n \text{exp}(-\mu).
\end{equation}
Questa distribuzione è molto spesso usata per modellare esperimenti di conteggio. In linea di principio, è valida solo se gli eventi individuali sono mutuamente indipendenti e il tasso è costante. Questo è il caso con i fotoni a raggi X che colpiscono un chip rivelatore, ma ad esempio non con il numero di persone che scendono da un treno nella stazione principale.

## Distribuzione Gaussiana

La distribuzione più comune (e quindi più abusata) è la distribuzione gaussiana (che è conosciuta anche con molti altri nomi, come la distribuzione **normale**). Si legge
\begin{equation}
\mathcal{N}(x | \mu, \sigma^2) = \frac{1}{\sqrt{2 \pi \sigma^2}} \text{exp}\left\{-\frac{1}{2} \left( \frac{x-\mu}{\sigma}\right)^2 \right\}.
\end{equation}

## Teorema del Limite Centrale

Sia $x_i\sim p(x)$, estratto da una distribuzione generica $p$, la media per un numero maggiore $N$ di tali estrazioni 
\begin{equation}
u = \frac{1}{N} \sum_{i=0}^{n-1} x_i \sim \mathcal{N}\left( u \Big| \langle x \rangle, \frac{1}{N}\text{Var}\left[x\right]\right),
\end{equation}
ossia segue una distribuzione gaussiana con valori medi **media** di x, 
\begin{equation}
\langle x \rangle = \int \text{d}x\, x p(x) \text{ or } \sum_k x_k p_k,
	\end{equation}
	e con $1/N$ volte la **varianza** di x, 
	\begin{equation}
\text{Var}\left[x\right]= \int \text{d}x\, (x-\langle x \rangle)^2p(x) \text{ or } \sum_k (x_k-\langle x \rangle)^2 p_k.
	\end{equation}
	
Il teorema del limite centrale è fondamentale per spiegare perché le variabili casuali che ricevono contributi da molti effetti indipendenti tendono ad essere distribuite come una gaussiana. Un buon esempio di ciò è l'altezza degli individui in una popolazione. Molti, piccoli fattori influenzano la differenza nella crescita dei membri della popolazione. Ma attenzione, molti processi sono dominati da pochi effetti, spesso fortemente (anti-)correlati, come la ricchezza degli individui in una popolazione. Se sei ricco, tendi ad avere più facilità nell'arricchirti ancora -- mi è stato detto. La distribuzione della ricchezza è molto lontana dalla gaussiana. 

Un altro contesto in cui il Teorema del Limite Centrale non vale è per distribuzioni che non hanno una media e/o varianza finite. Il più famoso di questi esempi è la 

## Distribuzione di Cauchy
\begin{equation}
	p(x| \alpha, \beta) = \frac{1}{\pi} \frac{\beta}{\beta^2 + (x-\alpha)^2}.
	\end{equation}

Sebbene questa distribuzione sia correttamente normalizzata e mostri anche un bel picco a $x=\alpha$, entrambi gli integrali per la media e la varianza divergono. Infatti, le somme di variabili casuali di una distribuzione di Cauchy sono distribuite seguendo una distribuzione di Cauchy. Una distribuzione veramente patologica e disturbata.

## Distribuzione Lognormale:

La distribuzione lognormale è una distribuzione di probabilità di una variabile casuale il cui logaritmo è normalmente distribuito. La sua funzione di densità di probabilità (PDF) è data dalla seguente formula:

$f(x | \mu, \sigma) = \frac{1}{x \sigma \sqrt{2\pi}} \exp\left(-\frac{(\ln x - \mu)^2}{2\sigma^2}\right)$

## Numeri Pseudo Casuali
La seguente ricorsione fornisce numeri pseudo casuali :
\begin{equation}
\begin{split}
& x_{i+1} = (A x_i + C) \text{mod} M \text{ con} \
& x_0 = \texttt{SEED},
\end{split}
\end{equation}

per $A, C, M, \texttt{SEED} \in \mathbb{N}$. Infatti, $x_i / M \in (0, 1)$.

Si noti che questo algoritmo è periodico, a un certo punto si ottiene lo stesso numero di nuovo.

Possiamo regolare $A, C, M$ in modo che il periodo sia massimo per qualsiasi seme. Il periodo massimo sarà $M$, cioè ogni numero $0,, ..., M-1$ è stato visitato una volta. Questa condizione è raggiunta per $M$ una potenza di $2$, $C$ dispari, e $A-1$ divisibile per 4 (un'istanza del Teorema di Hull-Dobell).

Nelle implementazioni pratiche i semi possono essere scelti per assumere valori specifici in modo che la generazione di numeri casuali sia gestibile e riproducibile. Questo è fondamentale per tracciare i bug.

```python
def MCRNG(seed, A, C, M):
    
    x = seed

    numbers = [ ]
    
    while True:
        x = ( A*seed + C )%M
        
        if (x in numbers):
            break
        else:
            numbers.append(x)
            seed = x
        
    return numbers
        
```

## Convoluzione tra due Funzioni di Densità di Probabilità (PDF)

La convoluzione è un'operazione matematica fondamentale in teoria della probabilità, che descrive la distribuzione di probabilità risultante dalla somma di due variabili aleatorie indipendenti. È particolarmente utile nell'analisi di sistemi lineari e nel trattamento dei segnali.

### Definizione

Data due funzioni di densità di probabilità $f(x)$ e $g(x)$, la loro convoluzione è definita come:

$ (f * g)(t) = \int_{-\infty}^{\infty} f(\tau) g(t - \tau) d\tau $

Questa espressione calcola la somma ponderata di $f$ e $g$, dove il peso è dato dalla funzione $g$ traslata.

### Interpretazione

- **Somma di Variabili Indipendenti**: Se $X$ e $Y$ sono due variabili aleatorie indipendenti con densità $f(x)$ e $g(y)$ rispettivamente, allora la densità della loro somma $X+Y$ è data dalla convoluzione delle loro densità.

### Proprietà

- **Commutativa**: $f * g = g * f$
- **Associativa**: $f * (g * h) = (f * g) * h$
- **Distributiva**: $f * (g + h) = f * g + f * h$
- **Elemento Neutro**: La delta di Dirac $\delta(x)$ agisce come elemento neutro nella convoluzione.

## Esercizio 2

La massa in Materia Oscura (DM) di ammassi di galassie con massa totale e' descritta da una distribuzione lognormale:

$P(M_\mathrm{DM}| M_\mathrm{tot}) = \mathrm{Lognorm}(\mu_\mathrm{DM}^\mathrm{True}, \sigma_\mathrm{DM}^\mathrm{True})$

## Metodo dell'Inverse Transform Sampling

Il metodo dell'Inverse Transform Sampling è una tecnica utilizzata in statistica per generare campioni casuali da qualsiasi distribuzione di probabilità, dato che si conosca la sua funzione di distribuzione cumulativa (CDF). Questo metodo è particolarmente utile quando si ha a che fare con distribuzioni complesse per le quali non esistono metodi di campionamento diretti.

### Principio di Base

Il principio fondamentale dietro l'Inverse Transform Sampling è che se una variabile casuale $U$ è distribuita uniformemente tra 0 e 1, allora la variabile $X = F^{-1}(U)$ avrà una distribuzione di probabilità $F$, dove $F^{-1}$ è l'inversa della funzione di distribuzione cumulativa di $X$.

### Passaggi del Metodo

1. **Generare U**: Si inizia generando un numero casuale $U$ da una distribuzione uniforme tra 0 e 1.

2. **Applicare l'Inversa della CDF**: Si calcola l'inversa della CDF ($F^{-1}$) per il valore di $U$. Questo può essere fatto analiticamente o numericamente, a seconda della complessità della CDF.

3. **Ottenere il Campione**: Il valore ottenuto, $X = F^{-1}(U)$, è un campione dalla distribuzione desiderata.


```python
from scipy.integrate import cumtrapz

yarr = convoluzione_valori
xarr = tau
def inverse_transform_sampling(xarr, yarr, num_samples):
    # Calcola la CDF a partire dalla PDF
    cdf = cumtrapz(yarr,xarr)

    # Genera campioni uniformi
    samples_uniform = np.random.rand(num_samples)

    # Trova i valori di x corrispondenti ai campioni uniformi utilizzando la CDF inversa
    samples = np.interp(samples_uniform, cdf, xarr[1:])

    return samples

# Esempio di utilizzo
num_samples = 10000  # Numero di campioni desiderati
%time samples = inverse_transform_sampling(xarr, yarr, num_samples)

plt.hist(samples,bins=51,density=True)
plt.plot(xarr,yarr)
```

## Algoritmo del Rejection Sampling

Il Rejection Sampling, noto anche come metodo di accettazione-rifiuto, è una tecnica utilizzata per generare campioni da una distribuzione di probabilità complessa, utilizzando una distribuzione più semplice da cui è facile campionare.

### Principio di Base

L'idea di base del Rejection Sampling è quella di utilizzare una distribuzione di probabilità nota e semplice da cui campionare (chiamata distribuzione proposta) per generare campioni che approssimano una distribuzione di probabilità più complessa (chiamata distribuzione target).

### Passaggi dell'Algoritmo

1. **Scegliere una Distribuzione Proposta**: Scegliere una distribuzione proposta $q(x)$ da cui è facile campionare e che soddisfi la condizione $q(x) \geq k \cdot p(x)$ per ogni $x$, dove $p(x)$ è la distribuzione target e $k$ è una costante.

2. **Generare un Campione**: Generare un campione $x$ dalla distribuzione proposta $q(x)$.

3. **Calcolare il Rapporto di Accettazione**: Calcolare il rapporto di accettazione $r = \frac{p(x)}{k \cdot q(x)}$.

4. **Generare un Numero Casuale Uniforme**: Generare un numero casuale uniforme $u$ tra 0 e 1.

5. **Accettare o Rifiutare il Campione**: Se $u \leq r$, accettare $x$ come campione dalla distribuzione target; altrimenti, rifiutare $x$ e ripetere il processo.


```python
import numpy as np

def rejection_sampling(xarr, yarr, num_samples):
    # Calcola il massimo valore di yarr per definire un limite superiore
    y_max = np.max(yarr)

    # Inizializza un array vuoto per i campioni accettati
    samples = []
    rejected = []
    while len(samples) < num_samples:
        # Campiona un indice a caso da xarr
        idx = np.random.randint(0, len(xarr))
        # Estrai un campione dalla distribuzione proposta
        x_sample = xarr[idx]

        # Estrai un valore y a caso tra 0 e y_max
        y_sample = np.random.uniform(0, y_max)

        # Accetta il campione se y_sample è minore del valore della PDF in x_sample
        if y_sample < yarr[idx]:
            samples.append(x_sample)
        else:
            rejected.append(x_sample)

    return np.array(samples),np.array(rejected)

# Esempio di utilizzo
num_samples = 10000  # Numero di campioni desiderati
%time samples, rejected = rejection_sampling(xarr, yarr, num_samples)

plt.hist(samples,bins=51,density=True)
plt.plot(xarr,yarr)

print("Sampling rate: ", len(samples)*1./(len(rejected)+len(samples)))
```

## Metodo di Campionamento Metropolis-Hastings

Il Metropolis-Hastings è un algoritmo di campionamento Monte Carlo a catena di Markov (MCMC) utilizzato per ottenere una sequenza di campioni casuali da una distribuzione di probabilità per la quale il campionamento diretto è difficile. Questo metodo è particolarmente utile per campionare da distribuzioni ad alta dimensionalità.

### Principio di Base

L'algoritmo Metropolis-Hastings crea una catena di Markov in cui lo stato successivo dipende solo dallo stato corrente, non dalla storia dei passaggi precedenti. L'obiettivo è costruire una catena di Markov che abbia la distribuzione target desiderata come sua distribuzione stazionaria.

### Passaggi dell'Algoritmo

1. **Inizializzazione**: Scegliere un punto iniziale casuale $x_0$ dalla distribuzione target.

2. **Iterazione**: Per ogni iterazione $i$:
   - Generare un candidato $x'$ a partire dallo stato corrente $x_i$ utilizzando una distribuzione proposta $q(x'|x_i)$.
   - Calcolare il rapporto di accettazione $\alpha = \min\left(1, \frac{p(x')q(x_i|x')}{p(x_i)q(x'|x_i)}\right)$, dove $p(x)$ è la densità della distribuzione target.
   - Generare un numero casuale $u$ da una distribuzione uniforme tra 0 e 1.
   - Se $u \leq \alpha$, accettare il candidato e porre $x_{i+1} = x'$; altrimenti, rifiutare il candidato e porre $x_{i+1} = x_i$.

3. **Raccolta dei Campioni**: Dopo un periodo iniziale di "burn-in", i valori di $x_i$ sono considerati campioni dalla distribuzione target.




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# Funzione target interpolata
target_distribution = interp1d(xarr, yarr, kind='cubic', bounds_error=False, fill_value=0)

def proposal_distribution(x):
    return np.random.normal(x, 0.2)

# Algoritmo Metropolis-Hastings
def metropolis_hastings(target_func, proposal_func, initial_value, n_samples):
    samples = [initial_value]
    current = initial_value
    rejected = []
    for _ in range(n_samples):
        candidate = proposal_func(current)
        accept_ratio = target_func(candidate) / target_func(current)
        if np.random.rand() < accept_ratio:
            current = candidate
        else: 
            rejected.append(candidate)
        samples.append(current)
    return samples[1:],rejected  # Rimuove il valore iniziale

# Generazione dei campioni
%time samples, rejected = metropolis_hastings(target_distribution, proposal_distribution, -4, 10000)
print("Sampling rate: ", len(samples)*1./(len(rejected)+len(samples)))
# Visualizzazione dei campioni
plt.hist(samples, bins=30, density=True, alpha=0.5)
plt.plot(xarr, yarr, label='Distribuzione Target')
plt.title('Campioni Metropolis-Hastings')
plt.legend()
plt.show()

plt.figure(1)
plt.plot(samples,'.')
plt.title('Trace plot')
```

## Distribuzione Normale Multivariata

La **Distribuzione Normale Multivariata** è una generalizzazione della distribuzione normale univariata a dimensioni superiori. È fondamentale in statistica e viene utilizzata in vari campi delle scienze naturali e sociali.

### Definizione

Un vettore casuale $\mathbf{X} = (X_1, X_2, \ldots, X_n)$ ha una distribuzione normale multivariata *n*-dimensionale se ogni combinazione lineare delle sue componenti ha una distribuzione normale univariata. Questa distribuzione descrive un insieme di variabili casuali che possono essere correlate tra loro.

### Parametri

La distribuzione è definita da due parametri principali:

1. **Vettore delle Medie $\boldsymbol{\mu}$**: Un vettore $\boldsymbol{\mu} = (\mu_1, \mu_2, \ldots, \mu_n)$, dove $\mu_i$ è la media di $X_i$.
   
   $\mu_i = E[X_i]$

2. **Matrice di Covarianza $\boldsymbol{\Sigma}$**: Una matrice $\boldsymbol{\Sigma}$ che descrive le covarianze tra le variabili. Gli elementi sulla diagonale principale sono le varianze di ciascuna variabile, mentre gli elementi fuori dalla diagonale sono le covarianze.

   $\Sigma_{ij} = \text{Cov}(X_i, X_j)$

### Definizioni Matematiche

- **Media ($\mu$)**: La media è il valore atteso di una variabile casuale. Rappresenta la posizione centrale della distribuzione.

  $\mu = E[X]$

- **Varianza ($\sigma^2$)**: La varianza è una misura della dispersione dei valori di una variabile casuale rispetto alla sua media.

  $\sigma^2 = E[(X - \mu)^2]$

- **Covarianza**: La covarianza è una misura del grado di variazione congiunta di due variabili casuali. Se le variabili tendono a variare insieme, la covarianza è positiva; se variano in direzioni opposte, è negativa.

  $\text{Cov}(X, Y) = E[(X - \mu_X)(Y - \mu_Y)]$

- **Coefficiente di Correlazione ($\rho$)**: Il coefficiente di correlazione è una misura normalizzata della covarianza che varia tra -1 e 1. Indica il grado e la direzione della relazione lineare tra due variabili.

  $\rho = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y}$

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Standard deviations and correlation coefficient
sigma1 = 1
sigma2 = 1
rho = 0.5  # Correlation coefficient

# Define the target distribution: 2D multivariate normal with specified covariance
mean = [0, 0]  # Mean of the distribution
cov = [[sigma1**2, rho * sigma1 * sigma2], [rho * sigma1 * sigma2, sigma2**2]]  # Covariance matrix
target_dist = stats.multivariate_normal(mean, cov)

# Calcolo del valore massimo della distribuzione target
max_target_pdf = target_dist.pdf(mean)

# Define the limits for the flat proposal distribution
x_min, x_max = -5 * sigma1, 5 * sigma1
y_min, y_max = -5 * sigma2, 5 * sigma2

# Rejection sampling function
def rejection_sampling(num_samples):
    samples = []
    rejected = []
    while len(samples) < num_samples:
        # Sample from the flat proposal distribution
        x = np.array([np.random.uniform(x_min, x_max), np.random.uniform(y_min, y_max)])

        # Calculate the acceptance probability
        accept_prob = target_dist.pdf(x) / max_target_pdf

        # Accept or reject the sample
        if np.random.rand() < accept_prob:
            samples.append(x)
        else:
            rejected.append(x)
    return np.array(samples), np.array(rejected)

# Number of samples to draw
num_samples = 10000

# Perform rejection sampling
samples, rejected = rejection_sampling(num_samples)

# Plot the samples
plt.scatter(samples[:, 0], samples[:, 1], alpha=0.5)
plt.title('Samples da una Distribuzione 2D Normale Multivariata')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
print("Sampling rate: ", len(samples)*1./(len(rejected)+len(samples)))
```

# Analisi delle Componenti Principali (PCA)

L'**Analisi delle Componenti Principali** (PCA, dall'inglese *Principal Component Analysis*) è una tecnica statistica utilizzata per ridurre la dimensionalità dei dati, mantenendo il più possibile la varianza presente nel dataset originale. Questo è particolarmente utile in contesti di visualizzazione dei dati, analisi della dimensionalita' effettiva, data science e machine learning, per la riduzione del rumore, e per migliorare l'efficienza di altri algoritmi di apprendimento automatico.

## Concetti Fondamentali

- **Varianza e Covarianza**: La PCA cerca di massimizzare la varianza lungo nuovi assi (componenti principali) e si basa sulla matrice di covarianza dei dati.
  
  - La **varianza** misura quanto i valori di una variabile si discostano dalla loro media. Matematicamente, è definita come:
    $$
    \text{Var}(X) = \frac{1}{n}\sum_{i=1}^{n}(X_i - \mu)^2
    $$
    dove $\mu$ è la media di \( X \).

  - La **covarianza** misura come due variabili variano insieme. È definita come:
    $$
    \text{Cov}(X, Y) = \frac{1}{n}\sum_{i=1}^{n}(X_i - \mu_X)(Y_i - \mu_Y)
    $$
    dove $\mu_X$ e $\mu_Y$ sono le medie di \( X \) e \( Y \), rispettivamente.

- **Componenti Principali**: Sono le nuove direzioni (assi) lungo le quali i dati sono proiettati. Questi assi sono ortogonali tra loro e catturano la massima varianza possibile.

- **Autovettori e Autovalori**: Gli autovettori della matrice di covarianza rappresentano le direzioni (o componenti principali) dei nuovi assi, mentre gli autovalori indicano la quantità di varianza catturata da ciascun autovettore.

## Procedura della PCA

1. **Standardizzazione dei Dati**: I dati vengono standardizzati sottraendo la media e dividendo per la deviazione standard.

2. **Calcolo della Matrice di Covarianza**: Si calcola la matrice di covarianza dei dati standardizzati.

3. **Calcolo degli Autovettori e Autovalori**: Si trovano gli autovettori e gli autovalori della matrice di covarianza.

4. **Selezione delle Componenti Principali**: Si scelgono le componenti principali (autovettori) che catturano la maggior parte della varianza (basandosi sugli autovalori).

5. **Trasformazione dei Dati**: I dati originali vengono trasformati (proiettati) sulle nuove componenti principali.

## Applicazioni della PCA

- **Riduzione della Dimensionalità**: Utile per ridurre il numero di variabili di un dataset, mantenendo la maggior parte delle informazioni.
- **Visualizzazione dei Dati**: Aiuta a visualizzare dati ad alta dimensionalità in 2D o 3D.
- **Preprocessing per Algoritmi di Machine Learning**: Migliora le prestazioni e riduce il tempo di calcolo per algoritmi di apprendimento automatico.


```python
import numpy as np
import matplotlib.pyplot as plt


# Step 1: Standardize the Data
samples_std = (samples - np.mean(samples, axis=0)) / np.std(samples, axis=0)

# Step 2: Compute the Covariance Matrix
cov_matrix = np.cov(samples_std.T)

# Step 3: Compute Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Step 4: Sort the eigenvalues in descending order
sorted_index = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_index]
# Similarly sort the eigenvectors 
sorted_eigenvectors = eigenvectors[:, sorted_index]

# Step 5: Transform the Data
# Here, we choose to keep all components (for 2D data)
samples_pca = np.dot(samples_std, sorted_eigenvectors)

# Plotting the original and PCA-transformed data
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(samples[:, 0], samples[:, 1], alpha=0.5)
plt.title('Original Data')

plt.subplot(1, 2, 2)
plt.scatter(samples_pca[:, 0], samples_pca[:, 1], alpha=0.5)
plt.title('Data after PCA')

plt.show()
```

## Generare un epidemia (nel senso simulare)

Il numero di contagi giornaliero e' una realizzazione Poissoniana con valore di aspettazione dato da una funzione logistica:






```python
def logistica(t, k = 100., N0 = 1., t0=10.5, a=0.196):
    """Calcola la funzione di distribuzione  logistica."""
    return k / (1 + (1.*k/(N0) -1)*(np.exp(-a*(t - t0))))

def diff_logistica(t, k = 100., N0 = 1., t0=10.5, a=0.196):
    cumulativa = logistica(t, k = k, N0 = N0, t0=t0, a=a)
    diff = cumulativa - np.roll(cumulativa,1) 
    diff[0] = 0.
    return diff
```
