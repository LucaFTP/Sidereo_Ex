## Proprietà di base

• p(Ω) = 1, la probabilità dell'insieme di tutti gli eventi possibili Ω è uno.

Questa proprietà funge da normalizzazione.

## Qual e' l'Area/Volume?


In sintesi, le probabilità possono essere considerate come misure di
volume o area.
Questa interpretazione pone la domanda su quale volume le proba-
bilità misurino effettivamente. In generale, gli eventi non hanno
estensione nello spazio reale, che occuperebbero. Come tali,
le probabilità non misurano il volume fisico associato
agli eventi. Quando diciamo che l'evento A è più probabile dell'
evento B, l'estensione di A nello spazio del possibile è maggiore
di quella di B. Infatti, un evento con probabilità zero ha la
stessa estensione dell'insieme vuoto. Allo stesso modo, la probabilità
di tutti gli eventi possibili è uno, indicando che è certa.
In sintesi, il modo in cui interpretiamo la probabilità di un
evento A maggiore di B, è che siamo più sicuri
che A accada piuttosto che B accada. Molto banalmente, preferiremmo scommettere sull'evento più probabile A
piuttosto che sull'evento meno probabile B.

Gli eventi sono detti indipendenti, se p(Ei ∩ Ej) = 0.
Un insieme completo di eventi Ei sono eventi mutuamente indipendenti,
se Ei ∩ Ej = 0 per i ≠ j, e se la loro unione riempie l'intero
spazio degli eventi possibili, ∪ Ei = Ω. In questo caso

$\sum_i p(E_i) = 1$

# Principali Distribuzioni

## Uniforme


\begin{equation}
p(x) =\begin{cases}
		1, &\text{se } 0<x<1\\
        0, & \text{altrimenti}
	 \end{cases}
\end{equation}
 Una distribuzione molto semplice per cui tutte le variabili casuali tra $(0,1)$ sono ugualmente probabili.
 
## Bernoulli

Distribuzione di una variabile casuale che assume il valore $k=1$ (successo) con probabilità $p$ e il valore $k=0$ (insuccesso), altrimenti

\begin{equation}
 p(k) = p^k (1-p)^{1-k},
\end{equation}

che è solo una forma chiusa per dire $p(1)=p$, $p(0)= 1-p(1) = 1-p$.

## Binomiale

Probabilità di $k$ successi in $n$ tentativi con probabilità individuale di successo $p$,

\begin{equation}
p(k | n, p) =  \binom{n}{k}p^k (1-p)^{n-k},
\end{equation}
composta da a) il coefficiente binomiale $\binom{n}{k}$, che sono le possibili combinazioni di $k$ prelievi da $n$ elementi, b) $p^k$, la probabilità di $k$ successi, se ogni successo ha probabilità $p$, e c) $(1-p)^{n-k}$, la probabilità di $n-k$ insuccessi, se ogni insuccesso ha probabilità $1-p$.

## Geometrica
Considera un evento con probabilità di successo $p$. Qual è la probabilità di fallire $k$ volte prima che accada una volta? Seguendo la spiegazione della distribuzione binomiale, segue immediatamente che 
\begin{equation}
p(k|p) = (1-p)^k \,p
\end{equation}

Qual è la probabilità di successo entro $n$ tentativi o meno?

\begin{equation}
\begin{split}
p(k<n | p)& = p \sum_{k=0}^{n-1} (1-p)^k = p \frac{1-(1-p)^n}{1-(1-p)} = \\
& = 1 - (1-p)^n,
\end{split}
\end{equation}
 dove abbiamo usato l'espressione standard per una somma geometrica, $\sum_{k=0}^{n-1} r^k = (1-r^n) / (1-r)$.
 
## Variabili continue

Fino ad ora, abbiamo considerato solo probabilità di variabili casuali discrete. Vediamo come si potrebbe prendere il limite per variabili casuali continue e derivare la distribuzione esponenziale. 

Considera un piccolo intervallo di tempo $\Delta$, in ciascuno dei quali effettui un tentativo con probabilità di successo $p = \Delta \lambda$, con $\lambda$ costante. In un tempo $t$ avrai $n=t/\Delta$ tentativi. La probabilità di avere avuto un successo entro $n$ tentativi (cioè con $n$ o meno), è data dall'equazione sopra,
 \begin{equation}
 P\Big(k<n=\frac{t}{\Delta}\Big) = 1 - (1-p)^n = 1 - (1 - \lambda \Delta)^\frac{t}{\Delta}.
 \end{equation}
 Nel limite $\Delta \rightarrow 0$ (un tentativo in ogni istante di tempo) troviamo 
 \begin{equation}
 P(<t) = 1 - \text{exp}(-\lambda t),
 \end{equation}
 come una delle definizioni dell'esponenziale è $\text{exp}(x) = \lim_{n\rightarrow 0}(1+x n)^{1/n}$. Si noti che l'idea di un tentativo in ogni momento del tempo ha senso solo se ciascuno di questi tentativi ha una probabilità di successo infinitesimale, inversamente proporzionale alla durata infinitesimale del tentativo. Se i molti tentativi avessero avuto probabilità di successo finite, l'evento sarebbe accaduto con certezza in qualsiasi intervallo di tempo $\epsilon>0$. Questo ci lascia praticamente con un processo con un tasso costante $\lambda$ di occorrenza. Questo ci porta alla prossima distribuzione, e la prima continua: la
 
## Distribuzione esponenziale 
Probabilità che un evento con tasso $\lambda$ si verifichi al tempo $t$,
\begin{equation}
\text{Exp}(t|\lambda) = \frac{\text{d}}{\text{d}t} \left[ 1 - \text{exp}(-\lambda t) \right] = \lambda \text{exp}(-\lambda t).
\end{equation}
Questa è la distribuzione tipica per il tempo di decadimento radioattivo, anche se in quel contesto, spesso definiamo il tempo di dimezzamento $\tau_{1/2}$, dato da $P(<\tau_{1/2}|\lambda)=1/2$. Invertendo questa espressione si può facilmente trovare $\ln 2 = \lambda \tau_{1/2}$.


## Distribuzione di Poisson

Probabilità di contare $n$ eventi, se ci si aspettava che si verificassero $\mu$ eventi
  \begin{equation}
\text{Possion}(n|\mu) = \frac{1}{n!} \mu^n \text{exp}(-\mu).
\end{equation}
Questa distribuzione è molto spesso usata per modellare esperimenti di conteggio. In linea di principio, è valida solo se gli eventi individuali sono mutuamente indipendenti e il tasso è costante. Questo è il caso con i fotoni a raggi X che colpiscono un chip rivelatore, ma ad esempio non con il numero di persone che scendono da un treno nella stazione principale.

## Distribuzione Gaussiana

La distribuzione più comune (e quindi più abusata) è la distribuzione gaussiana (che è conosciuta anche con molti altri nomi, come la distribuzione **normale**). Si legge
\begin{equation}
\mathcal{N}(x | \mu, \sigma^2) = \frac{1}{\sqrt{2 \pi \sigma^2}} \text{exp}\left\{-\frac{1}{2} \left( \frac{x-\mu}{\sigma}\right)^2 \right\}.
\end{equation}
Questa distribuzione è stata scoperta dal Prof. Gauss e dalla sua squadra -- sì, aveva assunto diverse persone per lavorare come "calcolatori". Il Prof. Gauss era molto scrupoloso nel suo incarico, come ci si aspetterebbe da un funzionario statale prussiano, e quindi registrava sempre la previsione per una distanza inferita e la distanza effettiva una volta verificata. Poi annotava la differenza tra le due quantità, etichettandola $\Delta$. Alla fine, scoprì che queste deviazioni $\Delta$ seguono una distribuzione $p(\Delta)\sim \text{exp}\left\{ -\frac{1}{2}\frac{\Delta^2}{\sigma^2} \right\}$. Con sua grande angoscia, notò anche che più la campagna di misurazione si estendeva nelle campagne, più questo fattore $\sigma$ aumentava. E meno accurate diventavano le sue mappe.

## Teorema del Limite Centrale

Sia $x_i\sim p(x)$, estratto da una distribuzione generica $p$, la media per un numero maggiore $N$ di tali estrazioni 
\begin{equation}
u = \frac{1}{N} \sum_{i=0}^{n-1} x_i \sim \mathcal{N}\left( u \Big| \langle x \rangle, \frac{1}{N}\text{Var}\left[x\right]\right),
\end{equation}
ossia segue una distribuzione gaussiana con valori medi **media** di x, 
\begin{equation}
\langle x \rangle = \int \text{d}x\, x p(x) \text{ or } \sum_k x_k p_k,
	\end{equation}
	e con $1/N$ volte la **varianza** di x, 
	\begin{equation}
\text{Var}\left[x\right]= \int \text{d}x\, (x-\langle x \rangle)^2p(x) \text{ or } \sum_k (x_k-\langle x \rangle)^2 p_k.
	\end{equation}
	
Il teorema del limite centrale è fondamentale per spiegare perché le variabili casuali che ricevono contributi da molti effetti indipendenti tendono ad essere distribuite come una gaussiana. Un buon esempio di ciò è l'altezza degli individui in una popolazione. Molti, piccoli fattori influenzano la differenza nella crescita dei membri della popolazione. Ma attenzione, molti processi sono dominati da pochi effetti, spesso fortemente (anti-)correlati, come la ricchezza degli individui in una popolazione. Se sei ricco, tendi ad avere più facilità nell'arricchirti ancora -- mi è stato detto. La distribuzione della ricchezza è molto lontana dalla gaussiana. 

Un altro contesto in cui il Teorema del Limite Centrale non vale è per distribuzioni che non hanno una media e/o varianza finite. Il più famoso di questi esempi è la 

## Distribuzione di Cauchy
\begin{equation}
	p(x| \alpha, \beta) = \frac{1}{\pi} \frac{\beta}{\beta^2 + (x-\alpha)^2}.
	\end{equation}

Sebbene questa distribuzione sia correttamente normalizzata e mostri anche un bel picco a $x=\alpha$, entrambi gli integrali per la media e la varianza divergono. Infatti, le somme di variabili casuali di una distribuzione di Cauchy sono distribuite seguendo una distribuzione di Cauchy. Una distribuzione veramente patologica e disturbata.

## Numeri Pseudo Casuali
La seguente ricorsione fornisce numeri pseudo casuali :
\begin{equation}
\begin{split}
& x_{i+1} = (A x_i + C) \text{mod} M \text{ con} \
& x_0 = \texttt{SEED},
\end{split}
\end{equation}

per $A, C, M, \texttt{SEED} \in \mathbb{N}$. Infatti, $x_i / M \in (0, 1)$.

Si noti che questo algoritmo è periodico, a un certo punto si ottiene lo stesso numero di nuovo.

Possiamo regolare $A, C, M$ in modo che il periodo sia massimo per qualsiasi seme. Il periodo massimo sarà $M$, cioè ogni numero $0,, ..., M-1$ è stato visitato una volta. Questa condizione è raggiunta per $M$ una potenza di $2$, $C$ dispari, e $A-1$ divisibile per 4 (un'istanza del Teorema di Hull-Dobell).

Nelle implementazioni pratiche i semi possono essere scelti per assumere valori specifici in modo che la generazione di numeri casuali sia gestibile e riproducibile. Questo è fondamentale per tracciare i bug.

```python
def MCRNG(seed, A, C, M):
    
    x = seed

    numbers = [ ]
    
    while True:
        x = ( A*seed + C )%M
        
        if (x in numbers):
            break
        else:
            numbers.append(x)
            seed = x
        
    return numbers
        
```

```python
seed = 10
A = 36+1
C = 3
M = int(2**11)

start = tm.time()
x1 = np.array(MCRNG(seed, A, C, M))
print(tm.time()-start)
```

```python
# consider a multiple choice test with N=15 questions with four answers each, of which only one is correct. 
# How many correct answers will you get, if you randomly choose answers (p=1/4)

from scipy.special import comb 
# https://stackoverflow.com/questions/26560726/python-binomial-coefficient

N = 16
p = 0.25

def TakeTheTest(N, p):
    
    answers = rng.uniform(size=N)
    
    return np.sum(np.where(answers<p, 1, 0))
    
num = 2048

ks = np.array([TakeTheTest(N, p) for i in range(num)])
ks2 = rng.binomial(N, p, size=num)

ints = np.arange(15)

_ = plt.hist(ks, bins = ints-0.5, density=True)
_ = plt.hist(ks2, bins = ints-0.5, density=True, histtype='step')

prob = comb(N, ints[:-1])* p**ints[:-1] * (1-p)**(N-ints[:-1])

print(prob)

_ = plt.plot(ints[:-1], prob, '*')
```

```python


def Fails(p):
    
    count = 0
    
    while rng.random()>p:
        count += 1
    
    return count
```

```python
p = 0.3

ints = np.arange(20)

k_geo = np.array([Fails(p) for i in range(num)])
k_geo2 = rng.geometric(p, size=num)-1  # python defines it as numbers of trials to run until success,
# while we consider number of failures before success, which is one less

_ = plt.hist(k_geo, bins = ints-0.5, density=True)
_ = plt.hist(k_geo2, bins = ints-0.5, density=True, histtype='step')

prob =  p * (1-p)**ints[:-1]

print(prob)

_ = plt.plot(ints[:-1], prob, '*')
```
